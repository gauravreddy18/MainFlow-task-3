```python
import pandas as pd
import matplotlib.pyplot as plt
data = {
    'Month': ['January', 'February', 'March', 'April', 'May'],
    'Sales': [200, 300, 400, 500, 600],
    'Expenses': [150, 200, 250, 300, 350]}
df = pd.DataFrame(data)

# Creating the bar chart
plt.figure(figsize=(10, 5))
plt.bar(df['Month'], df['Sales'], color='#1f77b4', edgecolor='black', linewidth=1.2, label='Sales')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Sales', fontsize=14)
plt.title('Monthly Sales', fontsize=16)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```

```python
# Creating the line chart
plt.figure(figsize=(10, 5))
plt.plot(df['Month'], df['Sales'], marker='o', linestyle='-', color='#1f77b4', linewidth=2, markersize=8, label='Sales')
plt.plot(df['Month'], df['Expenses'], marker='s', linestyle='--', color='#ff7f0e', linewidth=2, markersize=8, label='Expenses')
plt.xlabel('Month', fontsize=14)
plt.ylabel('Amount', fontsize=14)
plt.title('Monthly Sales and Expenses', fontsize=16)
plt.grid(axis='both', linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
```
